# 2 $N$-body gravitational systems in 2d

Consider a system of $N$ bodies with masses $m_i = m$ for $i=1,\dots, N$ interacting via gravitational force between the bodies. The bodies move in circular orbits with radius $R$ with velocities $|\vec v_i(t)| = v$.

## a)

The orbits are given by
\begin{align}
    \vec r_i(t) &= R\cdot \begin{pmatrix}
        \cos\left(\frac{v}{R}\cdot t + \varphi_i\right) \\ \sin\left(\frac{v}{R}\cdot t + \varphi_i\right)
    \end{pmatrix},
    &
    \vec v_i(t) = \dot{\vec r_i}(t) &= v\cdot \begin{pmatrix}
        -\sin\left(\frac{v}{R}\cdot t + \varphi_i\right) \\ \cos\left(\frac{v}{R}\cdot t + \varphi_i\right)
    \end{pmatrix}
\end{align}
with $\varphi_i = (i-1) \frac{2\pi}{N}$

## b)

The gravitational force between two bodies is given by
\begin{equation}
    \vec F_{ij} = - \frac{Gm^2}{|\vec r_i(t) - \vec r_j(t)|^2} \cdot \frac{\vec r_i(t) - \vec r_j(t)}{|\vec r_i(t) - \vec r_j(t)|} = - Gm^2\cdot\frac{\vec r_i(t) - \vec r_j(t)}{|\vec r_i(t) - \vec r_j(t)|^3}.
\end{equation}


For $N$ bodies, the total force $\vec F_1$ acting on body 1 is given by

\begin{align}
    \vec F_1
        &= - Gm^2\cdot\sum\limits_{j=2}^N \frac{\vec r_1(t) - \vec r_j(t)}{|\vec r_1(t) - \vec r_j(t)|^3} 
        \\
        &\stackrel{t=0}{=} - Gm^2\cdot\sum\limits_{j=2}^N \frac{\begin{pmatrix} R-R\cdot \cos(\varphi_j) \\ 0-R\cdot\sin(\varphi_j) \end{pmatrix}}{\left|\begin{pmatrix} R-R\cdot \cos(\varphi_j) \\ 0-R\cdot\sin(\varphi_j) \end{pmatrix}\right|^3} 
        \\
        &= - \frac{Gm^2}{R^3}\cdot\sum\limits_{j=2}^N \frac{\begin{pmatrix}
            R \cdot (1-\cos(\varphi_j)) \\ -R\cdot\sin(\varphi_j) 
        \end{pmatrix}}{\left((1-\cos(\varphi_j))^2 + \sin^2(\varphi_j)\right)^\frac{3}{2}} 
        \\
        &\phantom{===}\text{with } (1-\cos(\varphi_j))^2 + \sin^2(\varphi_j) = 4\sin^2(\varphi_j/2) \\
        &= - \frac{Gm^2}{R^2}\cdot\sum\limits_{j=2}^N \frac{\begin{pmatrix}
            1-\cos(\varphi_j) \\ -\sin(\varphi_j) 
        \end{pmatrix}}{\left(4\sin^2(\varphi_j/2)\right)^\frac{3}{2}}
        \\
        &\phantom{===}\text{with } 1-\cos(\varphi_j) = 2\sin^2(\varphi_j/2)
        \\
        &= - \frac{Gm^2}{2^3 R^2}\cdot\sum\limits_{j=2}^N \frac{\begin{pmatrix}
            2\sin^2(\varphi_j/2) \\ -\sin(\varphi_j) 
        \end{pmatrix}}{|\sin(\varphi_j/2)|^3}
        \\
        &= - \frac{Gm^2}{2^3 R^2}\cdot\sum\limits_{j=2}^N \frac{\begin{pmatrix}
            2\sin^2(\varphi_j/2) \\ -\sin(\varphi_j) 
        \end{pmatrix}}{|\sin(\varphi_j/2)|^3} := \begin{pmatrix}
            F_x \\ F_y
        \end{pmatrix}
\end{align}

with
\begin{align}
    \underline{\underline{  F_x  }}
        &= - \frac{Gm^2}{2^3 R^2}\cdot\sum\limits_{j=2}^N \frac{2\sin^2(\varphi_j/2)}{|\sin(\varphi_j/2)|^3}
        \\
        &= - \frac{Gm^2}{4 R^2}\cdot\sum\limits_{j=2}^N \frac{1}{\sin(\varphi_j/2)}
        = \underline{\underline{  - \frac{Gm^2}{4 R^2}\cdot\sum\limits_{j=2}^N \frac{1}{\sin((j-1)\frac{\pi}{N})}  }}
\end{align}
and
\begin{align}
    F_y 
        &= - \frac{Gm^2}{2^3 R^2}\cdot\sum\limits_{j=2}^N \frac{-\sin(\varphi_j)}{\sin^3(\varphi_j/2)}
        \\
        &= \frac{Gm^2}{2^3 R^2}\cdot\sum\limits_{j=2}^N \frac{\sin((j-1) \frac{2\pi}{N})}{\sin^3((j-1)\frac{\pi}{N})}
        \\
        &= 0
\end{align}
because of the geometric symmetry argument, that for each body $i$ with $(\vec r_i)_y \neq 0$ there is a corresponding body $i'$ with $(\vec r_{i'})_y \hat{=} -(\vec r_i)_y$, that therefore compensates the contribution to the y-component of the force. (This can also be seen by splitting the sum into two canceling sums and for even $N$ an 0-term).

Therefore, it is

\begin{equation}
    \underline{\underline{  \vec F _1 = \begin{pmatrix}F_x \\ 0\end{pmatrix}  }}
\end{equation}
with $F_x$ like defined above.

## c)
With Newton's equation of motion
\begin{equation}
    \sum\limits_{j\neq 1} \vec F_{1j} = \vec F_1 \stackrel{!}{=} m_1 \ddot{\vec r}_1 = m\ddot{\vec r}_1
\end{equation}
and with 
\begin{align}
\ddot{\vec r}_1(t) = \dot{\vec v}(t) 
    &= -\frac{v^2}{R}\cdot \begin{pmatrix}
        \cos\left(\frac{v}{R}\cdot t\right) \\ \sin\left(\frac{v}{R}\cdot t\right)
    \end{pmatrix}
    \\
    &\stackrel{t=0}{=} -\frac{v^2}{R}\cdot \begin{pmatrix}
        1 \\ 0
    \end{pmatrix}
\end{align}
it is
\begin{align}
    m\cdot (\ddot{\vec r}_1)_x &\stackrel{!}{=} F_x \\
    \Leftrightarrow 
    -m\frac{v^2}{R} &= - \frac{Gm^2}{4 R^2}\cdot\sum\limits_{j=2}^N \frac{1}{\sin((j-1)\frac{\pi}{N})}
    \\
    \Rightarrow v^2 &= \underline{\underline{  \frac{Gm}{4 R}\cdot\sum\limits_{j=2}^N \frac{1}{\sin((j-1)\frac{\pi}{N})}  }}
\end{align}

## d) Simulation

At first, we have to reduce the set of $N$ coupled 2nd-order differential equations

\begin{equation}
    \sum\limits_{j\neq i} \vec F_{ij} = \vec F_i \stackrel{!}{=} m \ddot{\vec r_i}
\end{equation}

to a set of $2\cdot N$ vectorial (or $2\cdot 2\cdot N$ component-wise) coupled 1st-order differential equations:
\begin{align}
    \vec u_{i,1} &= \vec r_i 
        & \to \frac{d}{dt}\vec u_{i,1} &= \dot{\vec r}_i = \vec u_{i,2}
        & \left\{\begin{array}{ccc}
            u_{i,1}^x = r_i^x & \to & \frac{d}{dt}u_{i,1}^x = u_{i,2}^x \\
            u_{i,1}^y = r_i^y & \to & \frac{d}{dt}u_{i,1}^y = u_{i,2}^y
        \end{array}\right. \\
    \vec u_{i,2} &= \dot{\vec r}_i & \to \frac{d}{dt}\vec u_{i,2} &= \ddot{\vec r}_i = \frac{1}{m} \vec F_i
    & \left\{\begin{array}{ccc}
            u_{i,2}^x = \dot r_i^x & \to & \frac{d}{dt}u_{i,2}^x = \frac{1}{m} F_i^x \\
            u_{i,2}^y = \dot r_i^y & \to & \frac{d}{dt}u_{i,2}^y = \frac{1}{m} F_i^y
        \end{array}\right. \\
\end{align}

where $i=1,\dots, N$ and $x,y$ are the vector components. 

This gives us
\begin{align}
    \vec u &= \begin{pmatrix}
        u_{1,1}^x \\
        u_{1,2}^x \\
        u_{1,1}^y \\
        u_{1,2}^y \\
        \vdots \\
        u_{N,1}^x \\
        u_{N,2}^x \\
        u_{N,1}^y \\
        u_{N,2}^y
    \end{pmatrix}, &
    \vec f(\vec u, t) &= \begin{pmatrix}
        u_{1,2}^x \\
        \frac{1}{m}F_1^x \\
        u_{1,2}^y \\
        \frac{1}{m}F_1^y \\
        \vdots \\
        u_{N,2}^x \\
        \frac{1}{m}F_N^x \\
        u_{N,2}^y \\
        \frac{1}{m}F_N^y
    \end{pmatrix}
\end{align}
resp.
\begin{align}
    \vec u &= \begin{pmatrix}
        u_{1,1} \\
        u_{1,2} \\
        u_{2,1} \\
        u_{2,2} \\
        \vdots \\
        u_{N,1} \\
        u_{N,2}
    \end{pmatrix}, &
    \vec f(\vec u, t) &= \begin{pmatrix}
        u_{1,2} \\
        \frac{1}{m}F_1 \\
        u_{2,2} \\
        \frac{1}{m} F_2 \\
        \vdots \\
        u_{N,2} \\
        \frac{1}{m}F_N
    \end{pmatrix}
\end{align}
with
\begin{equation}
    \leadsto \boxed{\dot{\vec u}(t) = \vec f(\vec u, t)}
\end{equation}

```julia
using LinearAlgebra
using Plots
```

```julia
# Global constants
G = 1 # gravitational constant
m = 1 # masses
R = 1 # radius
N = 4 # number of bodies

# calculate the velocity using the equation from above
function v²()
    s = 0
    for j in 2:N
        s += 1/sin((j-1)*π/N)
    end
    return G * m/(4 * R) * s
end
v() = sqrt(v²()) # can be later overwritten to other values

# angles between the bodies
φ(i) = (i-1)*2π/N

# positions at t=0
r_start(i) = R .* [
    cos(φ(i)),
    sin(φ(i))
]
# velocities of the bodies
v_start(i) = v() * [
    -sin(φ(i)), cos(φ(i))
]

# gravitational force
Fᵢⱼ(r⃗ᵢ, r⃗ⱼ) = -G * m^2 * (r⃗ᵢ - r⃗ⱼ)/(norm(r⃗ᵢ - r⃗ⱼ))^3

# force on body i
function Fᵢ(rs, i)
    F = zeros(2)
    for j in 1:N
        if i == j
            # only sum for i != j
            continue
        end
        F += Fᵢⱼ(rs[:,i], rs[:,j])
    end
    return F
end

# reset u
function reset_u()
    u = [zeros(2) for _ in 1:2*N]

    # set positions
    u[1:2:2*N] = [r_start(i) for i in 1:N]
    #set velocities
    u[2:2:2*N] = [v_start(i) for i in 1:N]

    # convert to 2×2N matrix
    u = hcat(u...)
    return u
end

u = reset_u()

r_vals(u) = u[:, 1:2:2N]
v_vals(u) = u[:, 2:2:2N]

# function f
function f(u)
    f = zeros(2, 2N)
    # set d/dt u₁ = u₂
    f[:,1:2:2N] = u[:,2:2:2N]
    # set d/dt u₂ = 1/m F
    rs = r_vals(u)
    for i = 1:N
        f[:,2i] = 1/m * Fᵢ(rs, i)
    end
    return f
end
```

```julia
# not really used later, does not make simulation better
# set values lower threshold ϵ (eps()) to 0
function u_precision!(u)
    for i in 1:length(u)
        if abs(u[i]) < eps()
            u[i] = 0.0
        end
    end
    return u
end
```

```julia
# Implementation of the Runge-Kutta-Method
function runge_kutta_step_order_2(u)
    fi = f(u)
    a = f(u + Δt * fi)
    u_next = u + 0.5 * Δt * fi + 0.5 * Δt * a
    return u_next
end

function runge_kutta_step_order_4(u)
    c1 = Δt * f(u)
    c2 = Δt * f(u + 0.5 * c1)
    c3 = Δt * f(u + 0.5 * c2)
    c4 = Δt * f(u + c3)
    u_next = u + (c1 + 2.0 * c2 + 2.0 * c3 + c4)/6.0
    return u_next
end
```

```julia
# compare with theory
r_theory(t,i) = R .* [
    cos(sqrt(v²())/R * t + φ(i)),
    sin(sqrt(v²())/R * t + φ(i))
]
```

```julia
# simulate
N = 4
R = 2
m = 1

Δt = 0.1
T_period = 2π*R/v() # time theoretically needed for one circle at velocity v
NUMBER_CIRCLES = 5
NUMBER_STEPS = NUMBER_CIRCLES * T_period/Δt # number of time steps to complete NUMBER_CIRCLES (in theory)

u = reset_u()
#u_precision!(u)

scatter(r_vals(u)[1,:], r_vals(u)[2,:], aspect_ratio=:equal, color="black", markersize=3)
anim = @animate for i in 1:NUMBER_STEPS
    global u
    u = runge_kutta_step_order_4(u)
    #u_precision!(u)
    rs = r_vals(u)
    scatter!(rs[1,:], rs[2,:], aspect_ratio=:equal, legend=false, color="black", markersize=3)
    #scatter!(hcat(r_theory.(i*Δt, 1:N)...)[1,:], hcat(r_theory.(i*Δt, 1:N)...)[2,:], aspect_ratio=:equal)
    #quiver!(rs[1,:], rs[2,:], quiver=(Δt*v_vals(u)[1,:], Δt*v_vals(u)[2,:]))
    xlims!((-1.1R,1.1R))
    ylims!((-1.1R,1.1R))
end

savefig("2d_N$(N)_w_$(NUMBER_CIRCLES)_circles_final.png")
mp4(anim, "2d_N$(N)_w_$(NUMBER_CIRCLES)_circles.mp4", fps=15)
```
