# Implementation of Convay's Game of Life

## a)

```julia
# in this context it's okay to use global variables for the whole game, 
# as this notebook has no other functionalities than the game

# define the global game parameters
GRID_WIDTH = 20 # width of the grid
GRID_HEIGHT = 20 # height of the grid
SIMULATE_STEPS = 50 # number of simulation steps to calculate, t ∈ [1, SIMULATE_STEPS+1] as t="0"→1

# initialize/reset the grid
function empty_grid()
    global GRID
    GRID = zeros(Int, GRID_HEIGHT, GRID_WIDTH, SIMULATE_STEPS+1);
end

# value at tuple position
function grid_at(idx_tuple, t)
    return GRID[idx_tuple[1], idx_tuple[2], t]
end

### steps needed for generating the next simulation step
## 1. functions to manipulate coordinates with periodic boundaries
left(col) = (if col == 1 return GRID_WIDTH else return col - 1 end)
right(col) = (if col == GRID_WIDTH return 1 else return col + 1 end)
up(row) = (if row == 1 return GRID_HEIGHT else return row - 1 end)
down(row) = (if row == GRID_HEIGHT return 1 else return row + 1 end)

## 2. get neighbour coordinates of cell (i,j) # u(p)/d(own), l(eft)/r(ight)
# format (row, column)
function get_neighbours(i,j)
    return [(up(i), left(j)), (up(i), j), (up(i), right(j)), (i, left(j)), (i, right(j)), (down(i), left(j)), (down(i), j), (down(i), right(j))]
end

## 3. get number of alive cells in neighbourhood
n(i,j,t) = sum(grid_at.(get_neighbours(i,j),t))

## 4. ruleset for updates
function f(z::Int,n::Int)
    if (z == 0 && n == 3) || (z == 1 && (n == 2 || n == 3))
        return 1 # reproduction / nothing
    else
        return 0 # under-/overpopulation / nothing
    end
end

function f(i::Int,j::Int,t::Int)
    # current value
    z = grid_at((i,j), t)
    return f(z,n(i,j,t))
end

# calculate the next frame (t+1) from the current (t)
function next_frame(t)
    global GRID
    for i in 1:GRID_HEIGHT, j in 1:GRID_WIDTH
        GRID[i,j,t+1] = f(i,j,t)
    end
end

# whole simulation (without animation)
function simulate()
    for t in 1:SIMULATE_STEPS
        next_frame(t)
    end
end
```

```julia
using PyCall
anim = pyimport("matplotlib.animation")

using PyPlot
using Base64, ProgressMeter


pygui(false)
```

```julia
function showanim(filename)
    base64_video = base64encode(open(filename))
    display("text/html", """<video controls src="data:video/x-m4v;base64,$base64_video">""")
end

function init_fig(;figsize=(10,10))
    global fig
    fig = figure(figsize=figsize)
    grid(true)
    xticks(GRID_WIDTH-0.5:-1:0.5)
    yticks(0.5:GRID_HEIGHT-0.5)
    xlim((-0.5, GRID_WIDTH-0.5))
    ylim((GRID_HEIGHT-0.5, -0.5))
end

#fig = figure()

#writer = anim.FFMpegWriter(fps=10)
#filename = "test.mp4"
#writer.setup(fig, filename)

#@showprogress for i in 1:SIMULATE_STEPS
#    imshow(GRID[:,:,i], cmap="gray_r")
#    writer.grab_frame()
#end

#writer.finish()
#close(fig)

#showanim(filename)
```
