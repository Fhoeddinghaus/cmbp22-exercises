# Implementation of Rule $N$

$N \in [0,255]$

```julia
N = 30

## convert N into the ruleset/table/map
N₂_map(N) = digits(N, base=2, pad=8) # in order of 1,2,4,8,16,32,64,128
```

```julia
# get value of b' = f(a,b,c)
function f(N::Int, abc::String)
    abc = parse(Int, abc, base=2)
    return N₂_map(N)[abc + 1] # note: julia counts from 1 not from 0!
end
```

```julia
# get cell value at i from current configuration
# configuration z is vector with indices from 1 to M
function f(N::Int, z::Vector, i::Int)
    if (i < 1) || (i > length(z)) 
        throw(BoundsError)
        return nothing
    end
    
    # get current values of neighbours
    # previous (i-1)
    a = 0
    if i > 1
        # get a from configuration
        a = z[i-1]
    end
    # current (i)
    b = z[i]
    # next (i+1)
    c = 0
    if i < length(z)
        c = z[i+1]
    end
    # concat to make abc
    abc = string(a,b,c)
    return f(N, abc)
end
```

```julia
# calculate next configuration from current
function next_z(N::Int, z::Vector)
    z′ = zeros(Int, length(z))
    for i in 1:length(z)
        z′[i] = f(N, z, i)
    end
    return z′
end
```

## Start configuration

\begin{align}
    z_i(t=0) &= 0 & \forall i=1,\dots, 120\\
    z_{60}(t=0) &= 1
\end{align}

```julia
N = 30
z_start = zeros(Int,120)
z_start[60] = 1
;
```

```julia
# calculate 50 generations
zs = zeros(Int, 51, 120)
zs[1,:] = z_start

for i in 2:51
    zs[i,:] = next_z(N, zs[i-1,:])
end
```

```julia
using Colors, Plots, L
plotlyjs()
```

```julia
p = heatmap(
    1:1:120, # x-axis label
    0:1:50,  # y-axis label
    zs, # all configurations to plot
    color = reverse(cgrad(:greys)), # reverse the map (0=white, 1=black)
    yflip=true, # t=0 at top
    size=(1200/4*3,510/4*3), 
    legend=false,
    xlabel="i",
    ylabel="Time t"
)
```
